```python
# Boxplot e verificação de valores extremos
# df1[num].boxplot()

plt.boxplot(df1[['Lote', 'Qtd_Parte']])
plt.show()
```

```python
# Tratando valores extremos.

df1_num = df1

# Calcular o intervalo interquartil (IQR) para detectar outliers
def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

# Aplicar a função para cada coluna
df1_num = remove_outliers(df1_num, 'Qtd_Parte')
df1_num = remove_outliers(df1_num, 'Lote')

# Nova verificação de valores extremos
plt.figure(figsize = (12, 4))

plt.boxplot(df1_num[['Lote', 'Qtd_Parte']])
plt.title("Boxplot")
plt.ylabel('...')
plt.grid(True)
plt.tight_layout()

plt.show()
```

```python
# plt.scatter(df1_num['Qtd. Parte'], df1_num['DiaAbate'], color = 'blue')
# plt.title('Quantidade por Lote')
# plt.xlabel('Qtd. Parte')
# plt.ylabel('Dia Abate')
# plt.grid(True)

# # Executa o gráfico
# plt.show()

plt.figure(figsize= (12,4))
sns.scatterplot(x = 'Qtd_Parte', y = 'Mes_Abate', hue = 'Mes_Abate', data = df1_num, palette= 'viridis')

plt.title('Partes Abatidas por Dia agrupado')
plt.xlabel('Qtd. Partes Afetadas')
plt.ylabel('Mês Abate')
plt.legend(title = 'Clusters')
plt.grid(False)

plt.show()
```

```python
# Média de partes afetas por mês
meses = df1_num.groupby('Mes_Abate')['Qtd_Parte'].mean()

# Plotar gráfico de linha
plt.figure(figsize=(12, 4))
meses.plot(kind='line', marker='o')
plt.title('Quantidade Média de Partes Afetadas por Mês')
plt.xlabel('Mês do Abate')
plt.ylabel('Média Partes Afetadas')
plt.grid(True)
plt.show()
```
